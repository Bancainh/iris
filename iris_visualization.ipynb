### 1️ Display the first five rows of the dataset
Task: Show the first 5 records of the Iris dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
iris.head()
```

### 2️ Plot histograms for each feature

Task: Create histograms for all numeric columns.
Comment: Describe the general distribution of each feature (normal, skewed, multimodal, etc.).

```python
plt.figure(figsize=(12, 8))
for i, column in enumerate(iris.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=iris, x=column, hue='species', bins=15, kde=True, palette='Set2', edgecolor='black')
    plt.title(f'Histogram of {column} by Species')

plt.tight_layout()
plt.show()
```

- Phân bố của Sepal Langth & Sepal Width: Phân bố đa đỉnh do sự chồng lấn của 3 loài. Các phần bố riêng lẻ từng loài gần giống phân bố chuẩn
- Phần bố của Petal Length & Petal Width: Phân bố đa đỉnh và có sự tách biệt dõ dàng giữa các loài. Ở biểu đồ petal length thì là phân bố chuyển theo từng loài. Còn biểu đồ petal_width thì Setosa bị lệch phải, versicolor theo phân bố chuẩn còn virginica gần giống phân bố chuẩn.


### 3️ Create boxplots for each feature
Task: Draw boxplots for all numeric variables.
Comment: Identify any outliers and compare the spread of values between features.

```python
plt.figure(figsize=(12, 8))
for i, column in enumerate(iris.columns[:-1], 1): 
    plt.subplot(2, 2, i)
    sns.boxplot(x='species', y=column, data=iris, palette='Set2')
    plt.title(f'Boxplot of {column} by Species')

plt.tight_layout()
plt.show()
```

Sepal Width (Chiều rộng Đài hoa): Có nhiều điểm ngoại lai nhất (outliers).

Petal Length và Petal Width (Chiều dài & Chiều rộng Cánh hoa): Thể hiện sự tách biệt rõ ràng nhất giữa 3 loài (hộp ít chồng lấn).

Loài Setosa có sự phân tán (IQR) ở cánh hoa cực kỳ nhỏ (rất nhất quán).

Sepal Length và Sepal Width (Chiều dài & Chiều rộng Đài hoa): Thể hiện sự chồng lấn nhiều nhất giữa các loài (khó phân biệt bằng mắt thường).

### 13️ KDE plot for petal_length

Đỉnh (Peak): Một đỉnh rất cao và sắc nét, tập trung mạnh quanh giá trị $\mathbf{1.5}$ cm.

Versicolor (Màu cam):

-Hình dạng: Khá cân đối và gần giống phân bố chuẩn (normal). Đỉnh (Peak): Một đỉnh rõ ràng, tập trung quanh giá trị $\mathbf{4.4}$ cm.

Virginica (Màu tím nhạt):

-Hình dạng: Hơi trải dài hơn và có xu hướng lệch trái (left-skewed) nhẹ, nhưng nhìn chung gần giống phân bố chuẩn. Đỉnh (Peak): Một đỉnh rộng hơn, tập trung quanh giá trị $\mathbf{5.6}$ cm.

### 15️ Regression plot for petal_length vs petal_width
Task: Draw a regression plot between petal length and petal width.

### 20️ Combined subplots (2×2) for the four main features
Task: Create a 2×2 subplot grid showing histograms for each main feature.
Comment: Compare the overall distribution shapes — which features have similar ranges or distributions?

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 9))
fig.suptitle('Histograms of Iris Features by Species', fontsize=16, fontweight='bold')

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for ax, feature in zip(axes.flat, features):
    sns.histplot(
        data=iris, 
        x=feature, 
        hue='species', 
        bins=20, 
        kde=True, 
        palette='Set2', 
        multiple='stack',  
        ax=ax
    )
    ax.set_title(feature.replace('_', ' ').title())
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

Petal Width 0.0 - 2.5 cm (Nhỏ nhất)Đa đỉnh (Trimodal) rõ ràngTương tự Petal Length: Tách biệt rõ rệt giữa các loài.

Sepal Length 4.0 - 8.0 cmĐa đỉnh, gần chuẩn hơnTương tự Sepal Width: Chồng lấn nhiều giữa các loài.

Sepal Width 2.0 - 4.5 cmĐa đỉnh, gần chuẩn hơnTương tự Sepal Length: Chồng lấn nhiều giữa các loài.
